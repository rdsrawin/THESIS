```python
import os
import csv
import requests
from urllib.parse import quote

# 1) Configuration
CSV_PATH   = "THERMAL.csv"                                    # your CSV with a “DOI” column
OUTPUT_DIR = "xmlfiles"                                    # directory to save .xml files
API_KEY    = "e46dd526a567506b5b2a4a6dda9963cd"            # your Elsevier API key

os.makedirs(OUTPUT_DIR, exist_ok=True)

# 2) Read DOIs from the CSV
with open(CSV_PATH, newline="", encoding="utf-8") as f:
    dois = [
        row["DOI"].strip()
        for row in csv.DictReader(f)
        if row.get("DOI") and row["DOI"].strip()
    ]

if not dois:
    raise RuntimeError("No DOIs found in the CSV—check the file and its 'DOI' header.")

# 3) Download each article’s XML into OUTPUT_DIR/
def download_elsevier_xml(dois, api_key):
    base_url = "https://api.elsevier.com/content/article/doi/"
    headers  = {"X-ELS-APIKey": api_key, "Accept": "text/xml"}

    for doi in dois:
        # percent-encode the DOI to handle special characters
        encoded = quote(doi, safe="")
        url = f"{base_url}{encoded}?view=FULL"
        try:
            resp = requests.get(url, headers=headers, timeout=30)
            if resp.status_code == 200:
                fname = doi.replace("/", "_") + ".xml"
                fpath = os.path.join(OUTPUT_DIR, fname)
                with open(fpath, "w", encoding="utf-8") as xf:
                    xf.write(resp.text)
                print(f"✓ Saved: {fpath}")
            elif resp.status_code == 404:
                print(f"✗ {doi} → NOT FOUND (404)")
            else:
                print(f"✗ {doi} → HTTP {resp.status_code}: {resp.text[:150]}")
        except requests.RequestException as e:
            print(f"✗ {doi} → Error: {e}")

# 4) Execute
download_elsevier_xml(dois, API_KEY)
```

```python
from pathlib import Path
import pandas as pd
from lxml import etree   # pip install lxml

XML_DIR = Path('venv').with_name("xmlfiles")   # …/xmlfiles next to your script

def extract_paragraphs_from_dir(xml_dir: Path) -> pd.DataFrame:
    """Parse all Elsevier full-text XML files in `xml_dir` and return a DataFrame
       with columns: doc_id (file stem), para_id (0-based), text (paragraph)."""
    
    records = []
    parser  = etree.XMLParser(recover=True, huge_tree=True)  # tolerant parser
    
    for xml_path in xml_dir.glob("*.xml"):
        try:
            root   = etree.parse(str(xml_path), parser)      # lxml.etree _ElementTree
            doc_id = xml_path.stem                           # filename without .xml
            
            # grab every paragraph irrespective of namespace prefix
            for idx, para in enumerate(root.xpath('//*[local-name()="para"]')):
                text = " ".join(para.itertext()).strip()
                if text:                                     # skip empty paras
                    records.append({
                        "doc_id":  doc_id,
                        "para_id": idx,
                        "text":    text
                    })
        except Exception as e:
            print(f"⚠️  Parse error in {xml_path.name}: {e}")

    return pd.DataFrame.from_records(records)

# -------- usage --------------------------------------------------------------
df_paras = extract_paragraphs_from_dir(XML_DIR)
print(df_paras.head())          # preview
print("Total paragraphs:", len(df_paras))
```
